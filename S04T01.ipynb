# Exercici 3
Geolocalitza les IP's.

```python
# make a new data frame with th IP adresses
data_ip = data['IP'].value_counts().rename_axis('ip').reset_index(name='visist_id')
```

```python
data_ip
```

```python
from IPython.display import IFrame
documentation = IFrame(src='https://geopy.readthedocs.io/en/latest/', width=1000, height=500)
display(documentation)
```

```python
from ip2geotools.databases.noncommercial import DbIpCity

def ip_latitude(ip):
    try:
        order_ip = DbIpCity.get(ip,api_key='free')
        latitude = order_ip.latitude
        return latitude
    except:
        return np.nan 

def ip_city(ip):
    try:
        city_ip = DbIpCity.get(ip, api_key='free')
        city = city_ip.city
        return city
    except: 
        return np.nan

def ip_country(ip):
    try:
        country_ip = DbIpCity.get(ip, api_key='free')
        country = country_ip.country
        return country
    except: 
        return np.nan
    

data_ip['latitude']=data_ip.loc[:,'ip'].apply(lambda x: ip_latitude(x))
data_ip['city']=data_ip.loc[:,'ip'].apply(lambda x: ip_city(x))
data_ip['country']=data_ip.loc[:,'ip'].apply(lambda x: ip_country(x))
display(data_ip.head(data_ip.shape[0]))
```
